```python
# 1. Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
```

```python
# 2. Crear un dataset ficticio
datos = {
    'Edad': [18,19,20,18,21,22,19,20],
    'HorasEstudio': [2,5,4,1,6,7,3,5],
    'Asistencia': [80,95,90,70,98,100,85,92],
    'Promedio': [70,92,88,65,95,98,80,90]
}

df = pd.DataFrame(datos)
```

```python
# 3. Mostrar la tabla
print("Datos de estudiantes")
print(df)
```

```python
# 4. Gráfico de barras
plt.bar(df.index, df['Promedio'], color='cornflowerblue')
plt.title("Promedio por estudiante")
plt.xlabel("Estudiante")
plt.ylabel("Promedio")
plt.show()
```

```python
# 5. Gráfico de dispersión
plt.scatter(df['HorasEstudio'], df['Promedio'], color='green')
plt.title("Relación entre horas de estudio y promedio")
plt.xlabel("Horas de estudio")
plt.ylabel("Promedio")
plt.show()
```

```python
# 6. Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlGnBu")
plt.title("Mapa de calor de correlaciones")
plt.show()
```

```python
# 7. Árbol de decisión

# Crear variable objetivo
df['AltoRendimiento'] = (df['Promedio'] >= 90).astype(int)

X = df[['Edad','HorasEstudio','Asistencia']]
y = df['AltoRendimiento']

# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Entrenar modelo
modelo = DecisionTreeClassifier(max_depth=3)
modelo.fit(X_train, y_train)

# Mostrar árbol
plt.figure(figsize=(10,6))
plot_tree(
    modelo,
    feature_names=X.columns,
    class_names=["No","Sí"],
    filled=True
)
plt.title("Factores que influyen en un alto rendimiento académico")
plt.show()
```
