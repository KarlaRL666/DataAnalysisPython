```python
# Paso 1: Importar librerías necesarias
import pandas as pd # Para manejar datos en tablas
```

```python
# Paso 2: Crear un DataFrame con datos crudos (notas de estudiantes)
datos = {
 "nombre": ["Ana", "Luis", "María", "Pedro", "Juan"],
 "nota": [85, 90, None, 70, 95] # Hay un valor faltante (None)
}
df = pd.DataFrame(datos)
```

```python
# Paso 3: Mostrar los datos originales
print("Datos originales:")
print(df)
```

```python
# Paso 4: Limpiar los datos -> reemplazar valores faltantes con el promedio
promedio = df["nota"].mean(skipna=True) # Calcula el promedio ignorando None
df["nota"] = df["nota"].fillna(promedio)
```

```python
# Paso 5: Mostrar los datos ya limpios
print("\nDatos limpios:")
print(df)
```

```python
# Paso 6: Transformar los datos -> crear una columna con calificación en letras
def convertir_a_letra(nota):
  if nota >= 90:
    return "A"
  elif nota >= 80:
    return "B"
  elif nota >= 70:
    return "C"
  else:
    return "D"
df["calificacion"] = df["nota"].apply(convertir_a_letra)
```

```python
# Paso 7: Mostrar los datos transformados
print("\nDatos transformados:")
print(df)
```

```python
# Paso 8: Calcular estadísticas básicas
print("\nEstadísticas:")
print("Promedio de notas:", df["nota"].mean())
print("Nota máxima:", df["nota"].max())
print("Nota mínima:", df["nota"].min())
```
